## Succession

![Succession](https://wikimedia.org/api/rest_v1/media/math/render/svg/ff6ced056735eb771cfa1b1d9f24f1d1559a1d16)

```python
from sympy import symbols, plot, sympify, parse_expr

n = symbols('n')

expression_0 = 1 + n

print('1 + b')
```

```python
from numpy import arange

xlim = 10

[ '{}'.format(xi, expression_0.subs(n, xi)) for xi in arange(1, xlim + 1) ]
```

```python
ylim = expression_0.subs(n, xlim)

plot(expression_0, (n, 0, xlim), axis_center = (0, 0), xlim = (0, xlim), ylim = (0, ylim))
```

## Addition

![Addition](https://wikimedia.org/api/rest_v1/media/math/render/svg/55d836ee9d53a15216c3b84ba021f855d8976022)

```python
expression = 2 + n

print("2 + n")
```

```python
from numpy import arange

xlim = 10

[ '2 + {} = {}'.format(xi, expression.subs(n, xi)) for xi in arange(1, xlim + 1) ]
```

```python
ylim = expression.subs(n, xlim)

plot(expression, (n, 0, xlim), axis_center = (0, 0), xlim = (0, xlim), ylim = (0, ylim))
```

## Multiplication

![Multiplication](https://wikimedia.org/api/rest_v1/media/math/render/svg/ba1076a1f2b21643e107430269c03540bf7c11cf)

```python
expression_2 = 2 * n

expression_2
```

## Exponentiation

![Exponentiation](https://wikimedia.org/api/rest_v1/media/math/render/svg/9fd5fb3bc9a8371ec6156d5d2b25479e0ac5f64e)

```python
expression_3 = 2 ** n

expression_3
```

```python
[ '2^{} = {}'.format(xi, expression_3.subs(n, xi)) for xi in arange(1, xlim + 1) ]
```

```python
ylim = expression_3.subs(n, xlim)

plot(expression, expression_2, expression_3, (n, 0, xlim), axis_center = (0, 0), xlim = (0, xlim), ylim = (0, ylim))
```

```python
xlim = 6
ylim = expression_3.subs(n, xlim)

plot(expression, expression_2, expression_3, (n, 0, xlim), axis_center = (0, 0), xlim = (0, xlim), ylim = (0, ylim))
```
